```python
import pandas as pd
import os 
```

```python
df_orders = pd.read_csv('../staging_2/orders/orders.csv')
df_orders
```

```python
df_orders_items = pd.read_csv('../staging_2/orders_items/orders_items.csv')
df_orders_items
```

```python
df_products = pd.read_csv('../staging_2/products/products.csv')
df_products
```

```python
merged_df = pd.merge(df_orders, df_orders_items, on='order_id', how='inner')
final_merged_df = pd.merge(merged_df, df_products, on='product_id', how='inner')
```

```python
final_merged_df.columns.sort_values()
```

```python
final_merged_df.drop(columns=['list_price_y', 'timesamp_x', 'timesamp_y'], inplace=True)
```

```python
# Assuming 'name' contains the file name including the extension '.csv'
new_path = '../Information Mart/'

# Expand '~' to the user's home directory
new_path = os.path.expanduser(new_path)

# Construct the folder path with the file name
folder_path = os.path.join(new_path, 'orders_products_details')

# Create the folder
os.makedirs(folder_path, exist_ok=True)

# Construct the file path within the newly created folder
file_path = os.path.join(folder_path, 'orders_products_details.csv')

# Save brands_df to the file_path
final_merged_df.to_csv(file_path, index=False)
```
