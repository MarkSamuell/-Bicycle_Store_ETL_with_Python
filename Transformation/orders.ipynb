```python
import pandas as pd
import os
```

```python
df_orders = pd.read_csv('../staging_1/orders/orders.csv')
```

```python
# Convert 'required_date' and 'shipped_date' to datetime objects
df_orders['required_date'] = pd.to_datetime(df_orders['required_date'])
df_orders['shipped_date'] = pd.to_datetime(df_orders['shipped_date'])
```

```python
df_orders['late_delivery'] = (df_orders['required_date'] <= df_orders['shipped_date'])
```

```python
# Calculate latency days
df_orders['latency_days'] = (df_orders['required_date'] - df_orders['shipped_date']).dt.days.where(df_orders['required_date'] > df_orders['shipped_date'])
```

```python
# Create a new column 'weekend' as boolean type
df_orders['weekend'] = df_orders['shipped_date'].dt.dayofweek.isin([6, 0])  # 6 represents Sunday, 0 represents Monday
```

```python
# Assuming 'name' contains the file name including the extension '.csv'
new_path = '../staging_2/'

# Expand '~' to the user's home directory
new_path = os.path.expanduser(new_path)

# Construct the folder path with the file name
folder_path = os.path.join(new_path, 'orders')

# Create the folder
os.makedirs(folder_path, exist_ok=True)

# Construct the file path within the newly created folder
file_path = os.path.join(folder_path, 'orders.csv')

# Save brands_df to the file_path
df_orders.to_csv(file_path, index=False)
```
