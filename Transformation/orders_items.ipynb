```python
import pandas as pd
import os
```

```python
df_xchange = pd.read_csv('../staging_1/exchange/exchange.csv')
df_orders = pd.read_csv('../staging_1/order_items/order_items.csv')
```

```python
orders_df = pd.merge(df_orders, df_xchange, how='cross')
orders_df = orders_df.iloc[:, :-2]
orders_df['list_price_egp'] = round((orders_df['list_price'] * orders_df['EGP']), 2)
orders_df
```

```python
# Assuming 'name' contains the file name including the extension '.csv'
new_path = '../staging_2/'

# Expand '~' to the user's home directory
new_path = os.path.expanduser(new_path)

# Construct the folder path with the file name
folder_path = os.path.join(new_path, 'orders_items')

# Create the folder
os.makedirs(folder_path, exist_ok=True)

# Construct the file path within the newly created folder
file_path = os.path.join(folder_path, 'orders_items.csv')

# Save brands_df to the file_path
orders_df.to_csv(file_path, index=False)
```
