```python
import pandas as pd
import requests
# .get() method :: to send a request and recieve the replay

response = requests.get('https://openexchangerates.org/api/latest.json',\
                        params={'app_id':'a91ca668be7a4d42b77e05fbce252c70'})
response.status_code
```

```python
data = response.json()
data
```

```python
df = pd.json_normalize(data)
df.drop('disclaimer', axis=1, inplace=True)
```

```python
rename_function = lambda x: x[6:] if x.startswith('rates.') else x

df.rename(columns=rename_function, inplace=True)
```

```python
df
```

```python
df = df[['EGP', 'timestamp']]
df['source'] = 'API'
```

```python
df.to_csv('../landing/exchange.csv', index=False)
```
