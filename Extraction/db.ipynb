```python
import pandas as pd
import sqlalchemy as db
import psycopg2
# create connection with the database
con = db.create_engine('postgresql://postgres:123@localhost/bikestore')
```

```python
query = """
SELECT * 
FROM 
general."orders"

"""
orders = pd.read_sql(query, 
                           con=con) 

query = """
SELECT * 
FROM 
general."order_items"

"""
orders_items = pd.read_sql(query, 
                           con=con) 
```
