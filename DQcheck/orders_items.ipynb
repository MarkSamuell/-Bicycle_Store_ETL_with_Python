```python
import pandas as pd
import os
```

```python
path = '../landing/'
name = 'order_items.csv'
df = pd.read_csv(path+name)
df.info()
```

```python
duplicated_rows = df[df.duplicated()]

duplicated_rows 
```

```python
df_products = pd.read_csv(path+'products.csv')
df_products
```

```python
valid_item_ids = df_products['product_id'].astype(str).unique()
df = df[df['item_id'].isin(valid_item_ids)]
```

```python
df['item_id'].unique()
```

```python
df['item_id'] = df['item_id'].astype(int)
```

```python
df.info()
```

```python
# Filter rows where 'list_price' is within the range (200, 10000]
df = df[(df['list_price'] > 200) & (df['list_price'] <= 10000)]
```

```python
df.info()
```

```python
# Assuming 'name' contains the file name including the extension '.csv'
file_name = name[:-4]
new_path = '../staging_1/'

# Expand '~' to the user's home directory
new_path = os.path.expanduser(new_path)

# Construct the folder path with the file name
folder_path = os.path.join(new_path, file_name)

# Create the folder
os.makedirs(folder_path, exist_ok=True)

# Construct the file path within the newly created folder
file_path = os.path.join(folder_path, name)

# Save brands_df to the file_path
df.to_csv(file_path, index=False)
```
