```python
import pandas as pd
import os
```

```python
path = '../landing/'
name = 'orders.csv'
df = pd.read_csv(path+name)
df.info()
```

```python
# Create an empty DataFrame to store rejected rows
df_rejected_dates = pd.DataFrame(columns=df.columns)

date_format = "%Y-%m-%d"

for index, row in df.iterrows():
    rejected = False
    for col in ['order_date', 'required_date', 'shipped_date']:
        try:
            df.at[index, col] = pd.to_datetime(row[col], format=date_format)
        except pd.errors.OutOfBoundsDatetime:
            print(f"Skipped row {index} with out of bounds timestamp in column '{col}'")
            # Append rejected row to df_rejected_dates
            df_rejected_dates = df_rejected_dates.append(row, ignore_index=True)
            rejected = True
            break
    if rejected:
        # Drop rejected row from the original DataFrame
        df.drop(index, inplace=True)
```

```python
df_rejected_dates
```

```python
# Convert 'order_date', 'required_date', and 'shipped_date' columns to datetime
for col in ['order_date', 'required_date', 'shipped_date']:
    df[col] = pd.to_datetime(df[col], format=date_format)
```

```python
df.info()
```

```python
# Calculate the average number of days between 'order_date' and 'shipped_date'
avg_days_between = (df['shipped_date'] - df['order_date']).mean().days

# Fill missing values in 'shipped_date' column with 'order_date' + average days
df['shipped_date'].fillna(df['order_date'] + pd.Timedelta(days=avg_days_between), inplace=True)
```

```python
# Filter rows where 'order_date' falls within the range (2000-01-01 to 2024-12-31)
filtered_df = df[(df['order_date'] >= '2000-01-01') & (df['order_date'] <= '2024-12-31')]
```

```python
# Assuming 'name' contains the file name including the extension '.csv'
file_name = name[:-4]
new_path = '../staging_1/'

# Expand '~' to the user's home directory
new_path = os.path.expanduser(new_path)

# Construct the folder path with the file name
folder_path = os.path.join(new_path, file_name)

# Create the folder
os.makedirs(folder_path, exist_ok=True)

# Construct the file path within the newly created folder
file_path = os.path.join(folder_path, name)

# Save brands_df to the file_path
df.to_csv(file_path, index=False)
```
